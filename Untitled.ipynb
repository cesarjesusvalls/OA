```python
#https://en.wikipedia.org/wiki/Pontecorvo%E2%80%93Maki%E2%80%93Nakagawa%E2%80%93Sakata_matrix
#http://www.nu-fit.org/?q=node/166

t12 = math.radians(33.62)
t23 = math.radians(47.2)
t13 = math.radians(8.54)
dcp = math.radians(234)

c23 = cos(t23)
s23 = sin(t23)
c13 = cos(t13)
s13 = sin(t13)
c12 = cos(t12)
s12 = sin(t12)
e_dcp_p = cmath.exp(1j * dcp)
e_dcp_n = cmath.exp(-1j * dcp)

A = [[1, 0, 0], 
    [0, c23, s23],
    [0, -s23, c23]]

B = [[c13, 0, s13*e_dcp_n], 
    [0, 1, 0],
    [-s13*e_dcp_p, 0, c13]]

C = [[c12, s12, 0], 
    [-s12, c12, 0],
    [0, 0, 1]]

PMNS = np.matmul(np.matmul(A,B),C)
PMNS_conjugate = np.conjugate(PMNS)
PMNS_modul = np.sqrt(PMNS_conjugate*PMNS)
```

```python
def DmNO(i,j):
    if i==3 and j==2:
        return 2.494E-3
    if i==3 and j==1:
        return 7.4E-5+2.494E-3
    if i==2 and j== 1:
        return 7.4E-5
```

```python
#https://arxiv.org/pdf/1710.00715.pdf
def getOA(a,b,L,E,nu=True):
    delta_dirac = 1 if a==b else 0
    prob_a = 0
    prob_b = 0
    for i in [0,1,2]:
        for j in [0,1,2]:
            if j>i:
                prob_a += np.real(PMNS[a-1][i]*PMNS_conjugate[b-1][i]*PMNS_conjugate[a-1][j]*PMNS[b-1][j])*(sin(1.27*DmNO(j+1,i+1)*L/(E))**2)
                prob_b += np.imag(PMNS[a-1][i]*PMNS_conjugate[b-1][i]*PMNS_conjugate[a-1][j]*PMNS[b-1][j])*sin(2*1.27*DmNO(j+1,i+1)*L/(E))
                
    return delta_dirac -4*prob_a+2*prob_b
```

```python
x = np.linspace(0, 500, 500, endpoint=True)
mu_e   = [getOA(2,1,l,0.6) for l in x]
mu_mu  = [getOA(2,2,l,0.6) for l in x]
mu_tau = [getOA(2,3,l,0.6) for l in x]
tot = [sum(i) for i in zip(mu_e,mu_mu,mu_tau)]

plt.plot(x, mu_e)
plt.plot(x, mu_mu)
plt.plot(x, mu_tau)
#plt.plot(x, tot)
```
